```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras import layers
from keras.utils import to_categorical
from keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

x → imagens, y → labels

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
classification = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
print(x_train.shape) # 50k imagens, 32x32, 3 camadas de profundidade (R, G, B)
print(y_train.shape) # 50k labels
```

```python
index = 333
print(classification[y_train[index][0]])
img = plt.imshow(x_train[index])
```

```python
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
print(y_train)
print(y_train_one_hot)
```

Normalização dos pixels (0...255 para 0...1)

```python
x_train = x_train / 255
x_test = x_test / 255
```

Building CNN

```python
model = Sequential()
model.add(Conv2D(32, (5, 5), activation='relu', input_shape=(32,32,3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(500, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(250, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

categorical_crossentropy: função de perda para classificação multi-class

Adam é um método de gradiente descendente estocástico

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train_one_hot, 
           batch_size=256, epochs=10, validation_split=0.3 )
```

```python
new_image = plt.imread("img2.jpg")
img = plt.imshow(new_image)
```

Ajustar a imagem para 32x32

```python
from skimage.transform import resize
resized_image = resize(new_image, (32,32,3))
img = plt.imshow(resized_image)
```

```python
predictions = model.predict(np.array( [resized_image] ))
```

```python
first_element = np.flip(np.sort(predictions))[0][0]
index = 0

for z in predictions[0]:
    if z == first_element:
        break;
    index += 1
    
print('RESULT:', classification[index])
```
